```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = stopwords.words('english')
stop_words = set(stop_words)
stop_words = stop_words - {'no', 'nor', 'not','never','neither'}
# print(stop_words)
```

```python
%%time
# data cleaning function
import pandas as pd
import re

keep_two_letter_words =  {'bp','rx','id','gi','iv','dx','hr','pr'}

def clean_text(text):
    if pd.isna(text):
        return ""

    text = re.sub(r'\bempty\b', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    words = text.split()    
    # Remove stopwords len-2  words except keep_two_letter_words
    words = [word for word in words if word not in stop_words and
             (len(word) > 2 or word in keep_two_letter_words)]      
    # Remove repeated character words like "aa", "bbb", etc.
    pattern = re.compile(r'^([a-z])\1{1,}$')
    words = [word for word in words if not pattern.match(word)]    
    return ' '.join(words)
```

```python
%%time
df_1['n_processed'] = df_1['n_processed'].apply(clean_text)
```

```python
# corrected 
import os
from sklearn.base import clone

def train_label_model(i, X, Y, base_clf, label_codes, skip_codes=None):
    icd_code = label_codes[i]
    save_path = f"Models/Individual_Models/{icd_code}.joblib"

    # Default empty skip list
    if skip_codes is None:
        skip_codes = []

    # Skip if already trained or in skip list
    if os.path.exists(save_path) or icd_code in skip_codes:
        return f"Model {icd_code} skipped"

    # Check if both classes exist in Y[:, i]
    if len(np.unique(Y[:, i])) < 2:
        return f"Model {icd_code} skipped (only one class present)"

    start_time = time.time()
    model = clone(base_clf)
    model.fit(X, Y[:, i])
    joblib.dump(model, save_path, compress=3)

    end_time = time.time()
    duration = end_time - start_time

    # Save training log
    with open(f"Models/Individual_Models/log_{icd_code}.txt", "w") as f:
        f.write(f"Model {icd_code} trained in {duration:.2f} seconds\n")

    return f"Model {icd_code} trained" 
```

```python
import pandas as pd

# Define probability brackets
brackets = [
    (0.95, 1.01, '>0.95'),
    (0.9, 0.95, '0.900.95'),
    (0.8, 0.9, '0.800.90'),
    (0.7, 0.8, '0.700.80'),
    (0.6, 0.7, '0.600.70'),
    (0.5, 0.6, '0.500.60'),
    (0.4, 0.5, '0.400.50'),
    (0.28, 0.4, '0.280.40')
    # ,(0.28, 0.3, '0.280.30')
]

# Store results
results = []
test_data_count = len(resultdf)
total_dataset_count = len(resultdf_exc_other)
 
# Loop over brackets
for lower, upper, label in brackets:
    filtered_df = resultdf_exc_other[
        (resultdf_exc_other['avg_predicted_probability'] >= lower) &
        (resultdf_exc_other['avg_predicted_probability'] < upper)
    ]
    
    total_encounters = len(filtered_df)
    if total_encounters == 0:
        results.append({
            'Bracket': label,
            'Total_encounter': 0,
            'Encounter_counts': 0,
            'Coverage': 0,
            'accept': 0,
            'reject': 0,
            'add': 0,
            'Code Precision': 0,
            'Code Recall': 0,
            'Code F1': 0,
            'Encounter Precision': 0,
            'Encounter Recall': 0,
            'Encounter F1': 0
        })
        continue

    # Code-level stats
    total_accepted = filtered_df['accepted_codes_count'].sum()
    total_rejected = filtered_df['rejected_codes_count'].sum()
    total_missed = filtered_df['missed_codes_count'].sum()

    if (total_accepted + total_rejected) > 0:
        precision = total_accepted / (total_accepted + total_rejected)
    else:
        precision = 0.0

    if (total_accepted + total_missed) > 0:
        recall = total_accepted / (total_accepted + total_missed)
    else:
        recall = 0.0

    if (precision + recall) > 0:
        f1 = 2 * precision * recall / (precision + recall)
    else:
        f1 = 0.0

    # Encounter-level stats
    perfect_encounters = filtered_df[
        (filtered_df['rejected_codes_count'] == 0) & 
        (filtered_df['missed_codes_count'] == 0)
    ]
    fully_recalled_encounters = filtered_df[
        filtered_df['missed_codes_count'] == 0
    ]

    encounter_precision = len(perfect_encounters) / total_encounters
    encounter_recall = len(fully_recalled_encounters) / total_encounters

    if (encounter_precision + encounter_recall) > 0:
        encounter_f1 = 2 * encounter_precision * encounter_recall / (encounter_precision + encounter_recall)
    else:
        encounter_f1 = 0.0

    results.append({
        'Bracket': label,
        'Total_encounter': test_data_count,
        'Encounter_counts': total_encounters,  # assuming same
        'Coverage': total_encounters / total_dataset_count,
        'accept': total_accepted,
        'reject': total_rejected,
        'add': total_missed,
        'Code Precision': round(precision, 4),
        'Code Recall': round(recall, 4),
        'Code F1': round(f1, 4),
        'Encounter Precision': round(encounter_precision, 4),
        'Encounter Recall': round(encounter_recall, 4),
        'Encounter F1': round(encounter_f1, 4)
    })

# Create a summary DataFrame
summary_df = pd.DataFrame(results)
```

```python
import ast

def contains_other_safe(item):
    # Convert from string to Python object if needed
    if isinstance(item, str):
        try:
            item = ast.literal_eval(item)
        except Exception:
            return False  # skip invalid strings

    # Ensure it's a list
    if not isinstance(item, list):
        return False

    # Now check for 'other' in any form
    for element in item:
        if isinstance(element, str) and element == 'XYZ':   # her XYZ is used for other category codes
            return True
        elif isinstance(element, tuple) and element[0] == 'XYZ':
            return True
    return False

# Now filter the DataFrame
resultdf_exc_other = resultdf[~resultdf['predicted_codes_with_prob'].apply(contains_other_safe)]
```

```python
import pandas as pd

# Define probability brackets
brackets = [
    (0.95, 1.01, '>0.95'),
    (0.9, 0.95, '0.900.95'),
    (0.8, 0.9, '0.800.90'),
    (0.7, 0.8, '0.700.80'),
    (0.6, 0.7, '0.600.70'),
    (0.5, 0.6, '0.500.60'),
    (0.4, 0.5, '0.400.50'),
    (0.3, 0.4, '0.300.40')
    ,(0.2, 0.3, '0.200.30')
    ,(0.1,0.2,'0.1-0.20')
    ,(0.09,0.1,'0.09-0.10')
]

# Store results
results = []
test_data_count = len(resultdf)
total_dataset_count = len(resultdf_exc_other)
 
# Loop over brackets
for lower, upper, label in brackets:
    filtered_df = resultdf_exc_other[
        (resultdf_exc_other['avg_predicted_probability'] >= lower) &
        (resultdf_exc_other['avg_predicted_probability'] < upper)
    ]
    
    total_encounters = len(filtered_df)
    if total_encounters == 0:
        results.append({
            'Bracket': label,
            'Total_encounter': 0,
            'Encounter_counts': 0,
            'Coverage': 0,
            'accept': 0,
            'reject': 0,
            'add': 0,
            'Code Precision': 0,
            'Code Recall': 0,
            'Code F1': 0,
            'Encounter Precision': 0,
            'Encounter Recall': 0,
            'Encounter F1': 0
        })
        continue

    # Code-level stats
    total_accepted = filtered_df['accepted_codes_count'].sum()
    total_rejected = filtered_df['rejected_codes_count'].sum()
    total_missed = filtered_df['missed_codes_count'].sum()

    if (total_accepted + total_rejected) > 0:
        precision = total_accepted / (total_accepted + total_rejected)
    else:
        precision = 0.0

    if (total_accepted + total_missed) > 0:
        recall = total_accepted / (total_accepted + total_missed)
    else:
        recall = 0.0

    if (precision + recall) > 0:
        f1 = 2 * precision * recall / (precision + recall)
    else:
        f1 = 0.0

    # Encounter-level stats
    perfect_encounters = filtered_df[
        (filtered_df['rejected_codes_count'] == 0) & 
        (filtered_df['missed_codes_count'] == 0)
    ]
    fully_recalled_encounters = filtered_df[
        filtered_df['missed_codes_count'] == 0
    ]

    encounter_precision = len(perfect_encounters) / total_encounters
    encounter_recall = len(fully_recalled_encounters) / total_encounters

    if (encounter_precision + encounter_recall) > 0:
        encounter_f1 = 2 * encounter_precision * encounter_recall / (encounter_precision + encounter_recall)
    else:
        encounter_f1 = 0.0

    results.append({
        'Bracket': label,
        'Total_encounter': test_data_count,
        'Encounter_counts': total_encounters,  # assuming same
        'Coverage': total_encounters / total_dataset_count,
        'accept': total_accepted,
        'reject': total_rejected,
        'add': total_missed,
        'Code Precision': round(precision, 4),
        'Code Recall': round(recall, 4),
        'Code F1': round(f1, 4),
        'Encounter Precision': round(encounter_precision, 4),
        'Encounter Recall': round(encounter_recall, 4),
        # 'Encounter F1': round(encounter_f1, 4)
    })

# Create a summary DataFrame
summary_df = pd.DataFrame(results)
```
